```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data() # It will download the data which is already split in train and test
```

```python
X_train.shape # 60K images with resolution 28 x 28
```

```python
# in order to train the neutral network we need to convert the data between 0 and 1
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
# Now we will create Artificial Neural Model... Keras Sequential Model

model = Sequential()
```

```python
# will conver 28 x 28 to flatten i.e to 28 x 28  = 784
# will use Keras to Flatten

model.add(Flatten(input_shape = (28,28))) # convert 28 x 28 into 1D array
```

```python
model.add(Dense(128, activation = 'relu')) # first Dense Layer and activation function relu
model.add(Dense(10, activation = 'softmax'))  # 10 nodes as we have 10 x output nodes and when we have more than 1 nodes in output we will use softmax
```

```python
model.summary()
```

```python
# 784 x 128 + 128 = 100480
```

```python
# 128 * 10 + 10 = 1290
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer='Adam') # in sparse we donot need to do one hot encoding whereas in cat entropy we need to onehot encoding
```

```python
# Model Train
model.fit(X_train, y_train, epochs=10, validation_split = 0.2)
```

```python
# Predictions

y_prob = model.predict(X_test)
```

```python
y_prob
```

```python
y_prob.argmax(axis=1) # which value has the highest probability of a image
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
# In order to find the accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
